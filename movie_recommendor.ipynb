## these are a list of dictionaries containing genres....but the entire list is in a string

```python
import ast
```

```python
def get_genres(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i["name"])
    return genres
```

the i we will get from the columns will be a string so we have to convert it a list first which is done by ast

```python
merged['genres'] = merged['genres'].apply(get_genres)
```

```python
merged.head(1)
```

```python
merged['cast'][0]
```

```python
def get_top_cast(obj):
    cast = []
    obj = ast.literal_eval(obj)[:3]
    for i in obj:
        cast.append(i["name"])
    return cast
```

```python
merged['cast'] = merged['cast'].apply(get_top_cast)
```

```python
def get_crew(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            director.append(i["name"])
    return director
```

```python
merged['crew'] = merged['crew'].apply(get_crew)
```

```python
def get_keywords(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i["name"])
    return genres
```

### we dont want all these numbers as most commons words</br>
###  there are also things like seperate words for actor and actors

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    obj = []
    for i in text.split():
        obj.append(ps.stem(i))
    return " ".join(obj)
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
df.head(3)
```

```python
vectors = cv.fit_transform(df['tags'])
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
distances = cosine_similarity(vectors)
```

```python
distances.shape
```

```python
distances
```

the shape makes sense as it measure similarity of each point against all the points

```python
list(enumerate(distances[0]))
```

```python
sorted(list(enumerate(distances[0])),reverse=True,key=lambda x:x[1]) #just sorting one row as an example
```

we are doing the above because upon sorting we would lose the indexs so we use enumerate to store the indices and while sorting we specifyy key so that it will sort based on the second value(distance) and not the first(index)

```python
def recommend_movie(movie):
    indices = []
    index = df.loc[df['title'].str.contains(movie)].index[0]#to find the index of the moviewe are entering
    sort = sorted(list(enumerate(distances[index])),reverse=True,key=lambda x:x[1])
    indices=sort[1:6]#we are taking from 1 index since highest similarity will be with the same movie
    return df.iloc[list(map(lambda x:x[0],indices))]['title']
```

```python
recommend_movie("Avatar")
```
