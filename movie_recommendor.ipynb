```python
import ast
```

```python
def get_genres(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i["name"])
    return genres
```

the i we will get from the columns will be a string so we have to convert it a list first which is done by ast

```python
merged['genres'] = merged['genres'].apply(get_genres)
```

```python
merged.head(1)
```

```python
def get_top_cast(obj):
    cast = []
    obj = ast.literal_eval(obj)[:3]
    for i in obj:
        cast.append(i["name"])
    return cast
```

```python
merged['cast'] = merged['cast'].apply(get_top_cast)
```

```python
def get_crew(obj):
    director = []
    for i in ast.literal_eval(obj):
        if i["job"]=="Director":
            director.append(i["name"])
    return director
```

```python
merged['crew'] = merged['crew'].apply(get_crew)
```

```python
merged.head(1)
```

```python
merged['keywords'][0]
```

```python
def get_keywords(obj):
    genres = []
    for i in ast.literal_eval(obj):
        genres.append(i["name"])
    return genres
```

```python
merged['keywords'] = merged['keywords'].apply(get_keywords)
```

```python
merged.head(1)
```

```python
merged['overview'] = merged['overview'].str.split()
merged.head(1)
```

```python
merged['cast'] = merged['cast'].apply(lambda x:[i.replace(" ","") for i in x])
merged['crew'] = merged['crew'].apply(lambda x:[i.replace(" ","") for i in x])
merged['genres'] = merged['genres'].apply(lambda x:[i.replace(" ","") for i in x])
merged['keywords'] = merged['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
merged.head()
```

```python
merged = merged.drop(columns=['release_year'],axis=1)
```

```python
merged['tags'] = merged['overview'] + merged['genres'] + merged['keywords'] + merged['cast'] + merged['crew']
```

```python
df = pd.concat([merged['id'],merged['title'],merged['tags']],axis=1)
```

```python
df['tags'] = df['tags'].apply(lambda x: " ".join(x))
df['tags'][0]
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words='english',max_features=5000)
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
def stemming(text):
    obj = []
    for i in text.split():
        obj.append(ps.stem(i))
    return " ".join(obj)
```

```python
df['tags'] = df['tags'].apply(stemming)
```

```python
df.head(3)
```

```python
vectors = cv.fit_transform(df['tags'])
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
distances = cosine_similarity(vectors)
```

```python
distances.shape
```

```python
distances
```

the shape makes sense as it measure similarity of each point against all the points

```python
def recommend_movie(movie):
    indices = []
    index = df.loc[df['title'].str.contains(movie)].index[0]#to find the index of the moviewe are entering
    sort = sorted(list(enumerate(distances[index])),reverse=True,key=lambda x:x[1])
    indices=sort[1:6]#we are taking from 1 index since highest similarity will be with the same movie
    return df.iloc[list(map(lambda x:x[0],indices))]['title']
```

```python
recommend_movie("Avatar")
```
